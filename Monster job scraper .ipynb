```python
import requests
from bs4 import BeautifulSoup as soup
#Send get request:
URL = 'https://www.monster.fi/tyopaikat/haku/?cy=fi&stpage=1&page=10' # what is the url for all the jobs as what we did in TE?
page = requests.get(URL)
bsobj = soup(page.content, 'html.parser')
#Get Job title:
title = []
for header in bsobj.findAll('h2',{'class':'title'}):
    title.append(header.text.strip())
```

```python
company = []
for name in bsobj.findAll('div',{'class':'company'}):
  company.append(name.text.strip())
```

```python
company
```

```python
location = []
for locs in bsobj.findAll('div',{'class':'location'}):
  location.append(locs.text.strip())
location = location[1:]
```

```python
location
```

```python
#Create pandas data frame:
import pandas as pd
d1 = {'title':title,'company':company,'location':location}
df = pd.DataFrame.from_dict(d1)
```
